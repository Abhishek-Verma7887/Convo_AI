```python
import pandas as pd
import numpy as np
import nltk
from nltk import *
```

```python
nltk.download('punkt')
nltk.download('stopwords')
  
```

```python
from nltk.corpus import PlaintextCorpusReader
a=PlaintextCorpusReader('./dataset/','.*')
```

```python
a.fileids()
```

```python
a.sents()
```

```python
a.words()
```

```python
content=[]

for i in range(len(a.fileids())):
  content.append(''.join(a.raw(fileids=a.fileids()[i])))
```

```python
content
```

```python
normalized=[]

for i in range(len(content)):
  normalized.append(' '.join(word.lower() for word in content[i].split() if word.isalpha() ))
```

```python
normalized
```

```python
tokenized=[]
from nltk import word_tokenize

for i in range(len(normalized)):
  tokenized.append(word_tokenize(normalized[i]))
```

```python
tokenized
```

```python
from nltk.corpus import stopwords
stopword_list= stopwords.words(fileids='english')
```

```python
stopword_list
```

```python
filtered=[]
for i in range(len(tokenized)):
  filtered.append(' '.join(word for word in tokenized[i] if word not in stopword_list))
```

```python
filtered
```

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
final_corpa=[]

for i in range(len(filtered)):
  final_corpa.append(' '.join( ps.stem(word) for word in filtered[i].split() ))
```

```python
final_corpa
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cc=CountVectorizer()
X_1=cc.fit_transform(final_corpa)
print(X_1.toarray())
print(cc.get_feature_names())
```

```python
bow=pd.DataFrame(data=X_1.toarray(),columns=cc.get_feature_names(),index=a.fileids())
bow
```

```python
#changed word_count[word]+=1 to word_count[word]=1 to form co-occurence matrix
doc_count={}
for i in range(len(final_corpa)):
    word_count={}
    for word in final_corpa[i].split():
        if word not in word_count.keys():
            word_count[word]=0
        word_count[word]=1
    doc_count[a.fileids()[i]]=word_count
print(doc_count)
```

```python
bow_2_1hot=pd.DataFrame(doc_count)
bow_2_1hot.fillna(0,inplace=True)
bow_x=np.asarray(bow_2_1hot)
bow_2_1hot.head()
```

```python
co_occur=bow_x.dot(bow_x.T)
co_occur
new_coour=pd.DataFrame( co_occur,columns=cc.get_feature_names(),index=cc.get_feature_names())
new_coour.head(10)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf=TfidfVectorizer()
X_2= tf.fit_transform(final_corpa)
print(X_2.toarray())
print(tf.get_feature_names())
```
