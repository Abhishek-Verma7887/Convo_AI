```python
import pandas as pd
import numpy as np
import sklearn
import nltk
```

```python
from nltk.corpus import PlaintextCorpusReader
a=PlaintextCorpusReader('./dataset/','.*')
print(a.fileids())
```

```python
content=[]
for i in range(len(a.fileids())):
  content.append(''.join( a.raw(a.fileids()[i])))

content
```

```python
normalized=[]

for i in range(len(content)):
  normalized.append(' '.join(word.lower() for word in content[i].split() if word.isalpha() ))
normalized
```

```python
nltk.download('punkt')
```

```python
tokenized=[]
from nltk import word_tokenize
for i in range(len(normalized)):
  tokenized.append(word_tokenize(normalized[i]))
tokenized
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords

stopword_list= stopwords.words(fileids='english')
stopword_list
```

```python
filtered=[]
for i in range(len(tokenized)):
  filtered.append(' '.join( word for word in tokenized[i] if word not in stopword_list ))

filtered
```

```python
from nltk.stem import PorterStemmer

ps= PorterStemmer()
final=[]

for i in range(len(filtered)):
  final.append(' '.join([ps.stem(word) for word in filtered[i].split() ]))

final
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer()
X_1=cv.fit_transform(final)
print(X_1.toarray())
print(cv.get_feature_names())
```

```python
df1= pd.DataFrame(X_1.toarray(),columns=cv.get_feature_names(),index=a.fileids())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tt=TfidfVectorizer()
X_2= tt.fit_transform(final)
print(X_2.toarray())
print(tt.get_feature_names())
df2=pd.DataFrame(X_2.toarray(),columns=tt.get_feature_names(),index=a.fileids())
df2.head()
```
