```python
import numpy as np
import pandas as pd
```

```python
A= np.random.randint(1,100,200)
A.shape
```

```python
B=np.random.randint(2,40,200)
B.shape
```

```python
df=pd.DataFrame({1:A})
df2= pd.DataFrame({2:B})
df.head()
```

```python
df2.head()
```

```python
df=pd.concat([df,df2],axis=1)
df.head()
```

```python
new= df.cov()
arr=np.asarray(df)
print(np.cov(arr))
```

```python
X=df.iloc[:,0]
Y=df.iloc[:,1]
```

```python
import sklearn as sk
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
X_scaled = sc.fit_transform(pd.DataFrame(X))
```

```python
type(X_scaled)
```

```python
A1 = np.ones(X_scaled.shape[0])
A1=A1.reshape(len(A1),1)
```

```python
X_scaled= np.concatenate([A1,X_scaled],axis=1)
```

```python
X_scaled.shape
```

```python
# formula beta = (X*X.T).inv * X.T * Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled,Y.values,test_size=0.30,random_state=42)
```

```python
A = X_train.T.dot(X_train) 
```

```python
B= np.linalg.inv(A)
```

```python
C = B.dot(X_train.T)
```

```python
beta= C.dot(Y_train)
```

```python
beta
```

```python
predicted = X_test.dot(beta) 
```

```python
predicted
```

```python
from sklearn import metrics
```

```python
metrics.r2_score(predicted,Y_test)
```

```python
#beta with ridge regression
lamba =0.0001
A= X_train.T.dot(X_train)
A1 = np.identity( A.shape[0])
A= np.add(A,lamba*A1)
B= np.linalg.inv(A)
C= B.dot(X_train.T)
beta2 = C.dot(Y_train)

```

```python
beta2
```

```python
predicted_2 = X_test.dot(beta2)
metrics.r2_score(predicted_2,Y_test)
```
